```python
def remove_punctuation(word):
    return word.translate(word.maketrans('', '', string.punctuation))
lemmatizer = WordNetLemmatizer() 
def clean_text(tweet):
    printable = set(string.printable)
    list_word = []
    for word in re.split('[.\s,?!:;-]', tweet):
        word = word.lower()
        # Filter non-ASCII characters
        word = ''.join(
            filter(lambda x: x in printable, word))
        # Remove Punctuations
        word = remove_punctuation(word)
        word = lemmatizer.lemmatize(word)
        word = word.strip()
        if word == '' or word == None:
            continue
        list_word.append(word)
    return " ".join(list_word)
```

```python
train['cleaned_text'] = train['comment_text'].apply(lambda x : clean_text(x))
```

```python
w2c_model = Word2Vec(train['text_combined'], size=100, window=5, min_count=1, workers=4)
```

```python
train['embeddings'] = train['text_combined'].apply(lambda x : w2c_model.wv[x])
```

## Embedding NN:

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
import tensorflow_docs as tfdocs
import tensorflow_docs.plots
import tensorflow_docs.modeling
import tensorflow_hub as hub
```

```python
model_link = "https://tfhub.dev/google/tf2-preview/gnews-swivel-20dim/1"
```

```python
hub_layer = hub.KerasLayer(model_link, output_shape=[20], input_shape=[], 
                           dtype=tf.string, trainable=True)
```

```python
hub_layer(X[:3])
```

```python
def make_model():
    model = tf.keras.Sequential()
    model.add(hub_layer)
    model.add(tf.keras.layers.Dense(256, activation='relu'))
    model.add(tf.keras.layers.Dropout(0.2))
    model.add(tf.keras.layers.Dense(64, activation='relu'))
    model.add(tf.keras.layers.Dense(6, activation='softmax'))
    return model

model = make_model()
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.losses.BinaryCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
history = model.fit(X,y,
                    epochs=5,
                    batch_size=512,
                    verbose=1)
```

```python
history.on_batch_begin(1)
history.params
```
